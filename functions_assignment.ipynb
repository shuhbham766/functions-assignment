#  1. What is the difference between a function and a method in Python?
* a) Function
  - A function is a block of reusable code that performs a specific task.
  - It is defined using the def keyword.
  - Functions can exist independently and are not tied to any object or class.



```python
def greet(name):
    return f"Hello, {name}!"

print(greet("Shubham"))
```

- b)  Method
  - A method is a function that is associated with an object.
  - It is defined inside a class and operates on instances of that class (i.e., objects).
  - Methods take self as the first parameter to access instance data.

```python
class Person:
    def __init__(self, name):
        self.name = name

    def greet(self):  # This is a method
        return f"Hello, {self.name}!"

p = Person("Shubham")
print(p.greet())
```

```python
# 10 Recursive Functions

# Define a recursive function
def factorial(n):
    if n == 1:
        return 1
    else:
        return n * factorial(n - 1)

# Call the recursive function
print(factorial(5))
```

```python
# 11 Nested Functions

# Define a nested function
def outer_function(x):
    def inner_function(y):
        return x + y
    return inner_function

# Call the outer function
result = outer_function(10)
print(result(5))
```

```python
# 12 Generator Functions
# A generator function uses the yield keyword to produce a sequence of values lazily.
# It returns an iterator that can be iterated over using a loop or next().

# Define a generator function
def count_up_to(n):
    i = 1
    while i <= n:
        yield i
        i += 1

# Call the generator function
for num in count_up_to(5):
    print(num)


```

#  5. What are iterators in Python and how do they differ from iterables?
- Iterable
 - An iterable is any Python object capable of returning its elements one at a time.
 - It implements the __iter__() method, which returns an iterator.
    - Example :-	Lists, tuples, strings, dictionaries, sets
 -Key Characteristics of Iterables:
    - Can be looped over using a for loop.
    - Can be passed to the iter() function to get an iterator.

```python
my_list = [1, 2, 3]  # List is an iterable

# Using a for loop to iterate over the iterable
for item in my_list:
    print(item)
```

Iterators
 - An iterator is an object that represents a stream of data.

 - It implements two methods:

  - __iter__(): Returns the iterator object itself.

  - __next__(): Returns the next value from the iterator. When no more items are available, it raises the StopIteration exception.

- Iterators are created from iterables using the iter() function.

Key Characteristics of Iterators:
  - Maintain their state (i.e., they "remember" where they are during iteration).

  - Can only move forward (no resetting or going backward).

  - Exhaustible: Once all items are consumed, the iterator is exhausted and cannot be reused.

```python
my_list = [1, 2, 3]  # List is an iterable
my_iterator = iter(my_list)  # Create an iterator from the iterable

# Manually get the next item using the iterator
print(next(my_iterator))
print(next(my_iterator))
print(next(my_iterator))
print(next(my_iterator))  # Raises StopIteration exception
```

# 6. Explain the concept of generators in Python and how they are defined.
- A generator is a special type of iterator that lets you:
  - Produce values one at a time
  - Use less memory
  - Use the yield keyword instead of return
  - Get values lazily (only when you ask for them)

Unlike regular functions that compute all values at once and return them in a list, generators produce values one at a time and only when needed. This makes them memory-efficient and ideal for working with large datasets or infinite sequences.

```python
# Example: Fibonacci Generator

def fibonacci(n):
    a, b = 0, 1
    for i in range(n):
        yield a
        a, b = b, a + b

for num in fibonacci(5):
    print(num)
```

# 10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?
- map()
  - Purpose: Transforms each item in an iterable by applying a function.
  - Use Case: When you want to transform or modify all elements of a sequence.

```python
nums = [1, 2, 3, 4]
squared = list(map(lambda x: x * x, nums))
print(squared)
```

- filter()
  - Purpose: Filters items based on a condition (returns only items where the function returns True).
  - Use Case: When you want to select specific elements of a sequence based on a condition.(“Keep only what matches the condition.”)

```python
nums = [1, 2, 3, 4, 5]
even = list(filter(lambda x: x % 2 == 0, nums))
print(even)
```

- reduce() :- (from functools module)
  -  Purpose: Reduces an iterable to a single value by applying a function cumulatively.
  - Use Case: When you want to aggregate or combine elements into one result (e.g., summing, multiplying).
   
    (  “Combine everything into one result.”)

```python
from functools import reduce

nums = [1, 2, 3, 4]
total = reduce(lambda x, y: x + y, nums)
print(total)
```

#  1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.

```python
def sum_even_numbers(numbers):
    even_sum = 0
    for num in numbers:
        if num % 2 == 0:
            even_sum += num
    return even_sum
nums = [10, 5, 8, 3, 2, 7]
result = sum_even_numbers(nums)
print("Sum of even numbers:", result)
```

#  4. Write a Python function that checks if a given number is prime or not from 1 to 200.

```python
def is_prime(n):
    if n<2:
        return False

    for i in range(2,int(n**0.5)+1):
        if n%i==0:
            return False
    return True

prime_no=[ele for ele in range(1,201) if is_prime(ele)]
print(prime_no)
```

#  5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

```python
class FibonacciIterator:
    def __init__(self, n_terms):
        self.n_terms = n_terms  # total terms to generate
        self.count = 0          # how many terms have been generated
        self.a, self.b = 0, 1   # starting values of Fibonacci sequence

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.n_terms:
            raise StopIteration
        if self.count == 0:
            self.count += 1
            return 0
        elif self.count == 1:
            self.count += 1
            return 1
        else:
            self.a, self.b = self.b, self.a + self.b
            self.count += 1
            return self.a
fib = FibonacciIterator(10)

for num in fib:
    print(num, end=' ')
```

# 6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

```python
def powers_of_two(max_exponent):
    for exp in range(max_exponent + 1):
        yield 2**exp

for power in powers_of_two(5):
    print(power)

```

# 10. Create a Python program that uses `filter()` to remove all the vowels from a given string.

```python
def remove_vowels(s):
    vowels = 'aeiouAEIOU'
    return ''.join(filter(lambda ch: ch not in vowels, s))

# Example input
text = "Shubham Kumar is learning Python"

# Remove vowels
result = remove_vowels(text)

print("String without vowels:", result)
```
